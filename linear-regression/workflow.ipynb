### Feature Engineering

MLB statistics as provided don't include some of the most useful analytical metrics.
The following block derives each metric from first principles using standard baseball formulas.
Each is documented with its formula and source for cross-referencing.

```python
# Singles: not provided directly — derived from total hits minus extra-base hits
# Formula: 1B = H - 2B - 3B - HR
df2['1B'] = df2['H'] - df2['2B'] - df2['3B'] - df2['HR']

# Total Bases: weighted sum of hits by base value
# Formula (MLB.com): TB = 1B + (2×2B) + (3×3B) + (4×HR)
df2['TB'] = df2['1B'] + (2 * df2['2B']) + (3 * df2['3B']) + (4 * df2['HR'])

# Runs Created: estimates a player's offensive contribution in total runs
# Combines ability to get on base with ability to hit for extra bases
# Formula (MLB.com): RC = TB × (H + BB) / (AB + BB)
df2['RC'] = (df2['TB'] * (df2['H'] + df2['BB'])) / (df2['AB'] + df2['BB'])
df2['RC'] = df2['RC'].round()

# Batting Average: hits divided by at-bats
# Formula: B_AVG = H / AB
df2['B_AVG'] = (df2['H'] / df2['AB']).round(3)

# On-Base Percentage: how often a player reaches base
# Formula (baseball-reference): OBP = (H + BB + HBP) / (AB + BB + HBP + SF)
df2['OBP'] = ((df2['H'] + df2['BB'] + df2['HBP']) /
              (df2['AB'] + df2['BB'] + df2['HBP'] + df2['SF'])).round(3)

# Slugging Percentage: total bases per at-bat — values extra-base hits more than batting average
# Formula: SLG = (1B + 2×2B + 3×3B + 4×HR) / AB
df2['SLG'] = ((df2['1B'] + (df2['2B'] * 2) + (df2['3B'] * 3) + (df2['HR'] * 4)) / df2['AB']).round(3)

# On-Base Plus Slugging: combines OBP and SLG into a single offensive measure
df2['OPS'] = df2['OBP'] + df2['SLG']

# Isolated Power: measures raw power by counting only extra-base hits
# Formula: ISO = SLG - B_AVG (or equivalently: (2B + 2×3B + 3×HR) / AB)
df2['ISO'] = df2['SLG'] - df2['B_AVG']
df2['ISO_XBH'] = ((1 * df2['2B']) + (2 * df2['3B']) + (3 * df2['HR'])) / df2['AB']
df2['ISO_XBH'] = df2['ISO_XBH'].round(3)

# BABIP: batting average on balls in play — removes outcomes the defense can't affect (HR, K)
# Formula: BABIP = (H - HR) / (AB - SO - HR + SF)
df2['BABIP'] = ((df2['H'] - df2['HR']) / (df2['AB'] - df2['SO'] - df2['HR'] + df2['SF'])).round(3)

# Checkpoint: cross-reference one player's computed stats against an external source
# https://www.baseballprospectus.com/player/99903/mike-zunino/
df2.iloc[-1]
```

### Inspect the Target Variable

Runs Created (RC) is our target — what we're trying to predict.

The distribution is right-skewed with a long tail. This makes it harder for a linear model to learn from, and means large RC values have outsized influence on the loss function.

```python
sns.histplot(df2['RC'], bins=50)
plt.title('Distribution of Runs Created (RC) — raw')
plt.show()
```

The long tail means the model would spend most of its effort trying to fit the extremes.

A log transformation compresses the scale — large values shrink closer to small ones, making the distribution more normal and easier to model. We use `log1p` (log(x+1)) to safely handle any zeros.

```python
log_runscreated = np.log1p(df2['RC'])

sns.histplot(log_runscreated, bins=50)
plt.title('Distribution of Runs Created (RC) — log transformed')
plt.show()

# np.expm1 is the inverse operation — undoes log1p to get back to original scale
# We'll use this at evaluation time to report errors in real RC units
sns.histplot(np.expm1(log_runscreated), bins=25)
plt.title('Inverse transform check — should match raw distribution')
plt.show()
```

### Validation Framework

Split the data into three subsets:
- **Train (60%):** the model learns from this
- **Validation (20%):** used to tune and compare models without touching the test set
- **Test (20%):** held out entirely — only used for the final evaluation

This prevents the model from 'seeing' test data during development, which would give an overly optimistic picture of performance.

```python
from sklearn.model_selection import train_test_split

# First split: 80% full_train, 20% test
df_full_train, df_test = train_test_split(df2, test_size=0.2, random_state=1)

# Second split: divide full_train into 60% train and 20% validation
# test_size=0.25 of 80% = 20% of total
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)

# Reset indices so all splits are numbered from 0
df_train = df_train.reset_index(drop=True)
df_val   = df_val.reset_index(drop=True)
df_test  = df_test.reset_index(drop=True)

print(f'Train: {len(df_train)} | Val: {len(df_val)} | Test: {len(df_test)}')
```

```python
# Isolate the target variable before removing it from the features
# Keep original values (for human-readable RMSE) and log-transformed (for model training)
y_train_orig = df_train.RC.values
y_val_orig   = df_val.RC.values
y_test_orig  = df_test.RC.values

y_train = np.log1p(df_train.RC.values)
y_val   = np.log1p(df_val.RC.values)
y_test  = np.log1p(df_test.RC.values)

# Remove the target column from all splits — the model must not see it as a feature
del df_train['RC']
del df_val['RC']
del df_test['RC']
```

### Baseline Model

Before adding engineered features, train on the raw columns only. This gives us a reference RMSE to beat.

Rather than using sklearn's `LinearRegression`, this notebook implements the closed-form solution (normal equations) from scratch — which was useful for understanding what the model is actually computing.

```python
def train_linear_regression(X, y):
    """
    Closed-form solution for linear regression using the normal equations.
    Computes weights that minimize the sum of squared errors between predicted and actual y.

    Returns: (w0, w) — bias term and weight vector
    """
    # Prepend a column of ones to represent the bias term (intercept)
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    # Gram matrix: X^T · X
    XTX = X.T.dot(X)
    # Invert the Gram matrix
    XTX_inv = np.linalg.inv(XTX)
    # Normal equation: w = (X^T X)^-1 · X^T · y
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]


def rmse(y_true, y_pred):
    """Root Mean Squared Error — average prediction error, penalizing large mistakes more."""
    return np.sqrt(((y_pred - y_true) ** 2).mean())
```

```python
# Raw features only — no engineered stats
base = ['G', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI',
        'SB', 'CS', 'BB', 'SO', 'IBB', 'HBP', 'SH', 'SF', 'GIDP']

X_train = df_train[base].fillna(0).values
w0, w = train_linear_regression(X_train, y_train)

y_pred = w0 + X_train.dot(w)

print(f'Baseline Train RMSE: {rmse(y_train, y_pred):.4f}')

sns.histplot(y_pred,  alpha=.5, color='red',  bins=50, label='Predicted')
sns.histplot(y_train, alpha=.5, color='blue', bins=50, label='Actual')
plt.legend()
plt.title('Predicted vs Actual (log scale) — baseline')
plt.show()
```

### Validating the Model

The `prepare_X` function standardizes how we transform any dataframe into a feature matrix. Having one reproducible function means train, val, and test all go through identical preprocessing.

```python
def prepare_X(df):
    """Prepare feature matrix from a dataframe. Fills missing values with 0."""
    df_num = df[base].fillna(0)
    return df_num.values

X_train = prepare_X(df_train)
w0, w = train_linear_regression(X_train, y_train)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

print(f'Validation RMSE (base features): {rmse(y_val, y_pred):.4f}')
```

### Feature Engineering Experiment

Now incorporate the engineered stats — OBP, SLG, OPS, ISO, BABIP, etc. — to see if they improve validation RMSE.

```python
# Extended feature list including engineered baseball metrics
base = ['G', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI',
        'SB', 'CS', 'BB', 'SO', 'IBB', 'HBP', 'SH', 'SF', 'GIDP',
        '1B', 'TB', 'B_AVG', 'OBP', 'SLG', 'OPS', 'ISO', 'ISO_XBH', 'BABIP']

X_train = prepare_X(df_train)
w0, w = train_linear_regression(X_train, y_train)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

print(f'Validation RMSE (with engineered features): {rmse(y_val, y_pred):.4f}')

sns.histplot([y_pred, y_val], alpha=.50, bins=50, palette='magma')
plt.title('Predicted vs Actual (log scale) — with engineered features')
plt.show()
```

The model predicts with *greater* error after adding engineered features. This is likely because the engineered stats (OBP, SLG, OPS etc.) are derived from the same raw columns already in the feature set — introducing redundancy and correlation that the model has difficulty resolving without regularization.

This is exactly the problem regularization is designed to address.

### Categorical Variables

One candidate feature: whether a player played in the American League (AL) or National League (NL). Historically, the AL uses a Designated Hitter while the NL has pitchers bat — which might affect offensive production patterns.

We one-hot encode the league column and test whether it helps.

```python
def prepare_X(df):
    """Prepare feature matrix. Includes one-hot encoding for league (AL/NL)."""
    df = df.copy()
    features = base.copy()

    # One-hot encode league — AL=1 means American League, NL=1 means National League
    df['lgAL'] = (df.lgID == 'AL').astype('int')
    df['lgNL'] = (df.lgID == 'NL').astype('int')
    features += ['lgAL', 'lgNL']

    return df[features].fillna(0).values

X_train = prepare_X(df_train)
w0, w = train_linear_regression(X_train, y_train)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

print(f'Validation RMSE (with league encoding): {rmse(y_val, y_pred):.4f}')
```

League encoding didn't meaningfully reduce RMSE. The AL/NL distinction doesn't add enough signal to offset the added complexity. We'll drop it and keep the final `prepare_X` to the engineered numeric features only.

```python
# Final prepare_X — engineered numeric features, no categorical encoding
def prepare_X(df):
    df_num = df[base].fillna(0)
    return df_num.values
```

### Regularization

When engineered features made performance worse, the likely culprit was **multicollinearity** — highly correlated features (e.g. OBP and OPS, which is OBP+SLG) destabilize the normal equations and inflate coefficient magnitudes.

**Ridge regularization (L2)** adds a penalty to the loss function equal to the sum of squared coefficients. This discourages any single feature from dominating and shrinks unstable coefficients toward zero — without removing them entirely.

Mathematically, instead of solving `X^T X · w = X^T y`, we solve:
`(X^T X + rI) · w = X^T y`

where `r` is the regularization strength. The identity matrix `I` adds a small constant to each diagonal entry, making the system better conditioned.

```python
def train_linear_regression_reg(X, y, r=0.001):
    """
    Ridge regression (L2 regularization).
    r: regularization strength. Higher = more shrinkage, lower = closer to unregularized.
    """
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    # Add regularization term to diagonal — this is what makes it Ridge
    XTX = XTX + r * np.eye(XTX.shape[0])

    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
```

### Tuning the Regularization Parameter

The regularization strength `r` is a hyperparameter — it's not learned from data, we have to choose it. We sweep across a range and pick the value with the lowest validation RMSE.

```python
print(f"{'r':>10}  {'Val RMSE':>10}")
print('-' * 25)

for r in [0.0, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100]:
    X_train = prepare_X(df_train)
    w0, w = train_linear_regression_reg(X_train, y_train, r)
    X_val = prepare_X(df_val)
    y_pred = w0 + X_val.dot(w)
    score = rmse(y_val, y_pred)
    print(f'{r:>10}  {score:>10.4f}')
print('-' * 25)
```

```python
# r=0.01 gives the best validation RMSE — use this for the final model
r = 0.01

X_train = prepare_X(df_train)
w0, w = train_linear_regression_reg(X_train, y_train, r)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

print(f'Validation RMSE (Ridge, r={r}): {rmse(y_val, y_pred):.4f}')

sns.histplot([y_pred, y_val], alpha=.50, bins=50, palette='magma')
plt.legend(['Predicted', 'Actual'])
plt.title('Predicted vs Actual (log scale) — Ridge regularized')
plt.show()
```

## Final Model — Test Set Evaluation

With the best regularization value selected, retrain on the full training set (train + validation combined) to use all available labeled data before testing on the holdout.

```python
# Combine train and validation for final training
df_full_train = pd.concat([df_train, df_val]).reset_index(drop=True)
y_full_train  = np.concatenate([y_train, y_val])

X_full_train = prepare_X(df_full_train)
w0, w = train_linear_regression_reg(X_full_train, y_full_train, r=0.01)

# Evaluate on held-out test set
X_test = prepare_X(df_test)
y_pred = w0 + X_test.dot(w)

print(f'Test RMSE (log scale): {rmse(y_test, y_pred):.4f}')
print(f'Val  RMSE (log scale): {rmse(y_val, w0 + prepare_X(df_val).dot(w)):.4f}')
```

### Evaluation Metrics

RMSE gives us one view of model quality, but three metrics together tell a more complete story:

- **RMSE** — average error, but penalizes large mistakes heavily (outliers pull it up)
- **MAE** — average *absolute* error, treats all mistakes equally — easier to interpret in plain terms
- **R²** — the proportion of variance in Runs Created that the model explains. R²=1.0 is perfect; R²=0.0 means the model does no better than predicting the mean every time

RMSE and MAE are in log-scale units here. To interpret in actual Runs Created, we inverse-transform with `expm1`.

```python
from sklearn.metrics import mean_absolute_error, r2_score

# Predictions are in log scale — inverse transform to get actual RC units
y_pred_actual = np.expm1(y_pred)
y_test_actual = np.expm1(y_test)

test_rmse = rmse(y_test, y_pred)
test_mae  = mean_absolute_error(y_test_actual, y_pred_actual)
test_r2   = r2_score(y_test, y_pred)

print('── Test Set Metrics ───────────────────────────')
print(f'  RMSE (log scale):          {test_rmse:.4f}')
print(f'  MAE  (actual RC units):    {test_mae:.2f} runs')
print(f'  R²   (variance explained): {test_r2:.4f}')
print('───────────────────────────────────────────────')
```

```python
# Actual vs Predicted scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(y_test_actual, y_pred_actual, alpha=0.3, color='steelblue')
plt.plot([0, y_test_actual.max()], [0, y_test_actual.max()],
         color='tomato', linestyle='--', label='Perfect prediction')
plt.xlabel('Actual Runs Created')
plt.ylabel('Predicted Runs Created')
plt.title(f'Actual vs Predicted Runs Created  (R² = {test_r2:.3f})')
plt.legend()
plt.tight_layout()
plt.show()
```

## Checking the Model on a Single Player

A practical sanity check — pick one player from the test set, run them through the model, and compare to their actual Runs Created.

```python
# Pull one player from the test set
player = df_test.iloc[20].to_dict()

df_small = pd.DataFrame([player])
X_small  = prepare_X(df_small)

y_pred_player = w0 + X_small.dot(w)
predicted_rc  = np.expm1(y_pred_player[0])
actual_rc     = np.expm1(y_test[20])

print(f'Predicted Runs Created: {predicted_rc:.1f}')
print(f'Actual Runs Created:    {actual_rc:.1f}')
print(f'Difference:             {abs(predicted_rc - actual_rc):.1f} runs')
```

## Reflection

Working through the linear algebra behind linear regression — implementing the normal equations by hand — made the jump to regularization feel natural rather than arbitrary. When you see that Ridge regression is just adding a small constant to the diagonal of the Gram matrix, it demystifies what sklearn's `Ridge(alpha=...)` is doing under the hood.

The evaluation metrics added here (MAE and R²) complement RMSE by giving both a plain-English interpretation (MAE in actual runs) and a relative quality measure (R²) that's easier to communicate to someone who isn't looking at the model code.

The same feature engineering mindset used here — deriving meaningful metrics from raw data using domain knowledge — translated directly into the admissions prediction work, where engineered applicant features often outperformed raw application fields.
